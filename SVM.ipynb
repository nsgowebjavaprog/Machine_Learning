```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
x = iris.data
y = iris.target
names = iris.target_names
```

```python
x.shape
```

```python
y.shape
```

```python
# Classsifictions

df = pd.DataFrame(x,columns = iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].replace(to_replace= [0,1,2], value = ['setpsa', 'versicolor','virginica'])
```

```python
df
```

```python
import seaborn as sns
sns.pairplot(data = df, hue='species', palette = 'Set2')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.22, random_state = 1)
```

```python
print(x_train.shape, y_train.shape)
print(y_test.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'linear', random_state = 0)
svm.fit(x_train, y_train)
```

```python
pred = svm.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
final_svm = SVC(kernel = 'rbf',random_state = 0)
final_svm.fit(x_train, y_train)
final_pred = final_svm.predict(x_test)
```

```python
accuracy_score(y_test, final_pred)
```
